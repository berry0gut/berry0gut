```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(method="ffill", inplace=True)
```

```python
df['전출지별']== '서울특별시'
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리하기 .


mask= (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df[mask]
```

```python
df_seoul=df[mask]
```

```python
# df_seoul에서 '전출지별 ' 컬럼 삭제

df_seoul= df_seoul.drop('전출지별', axis= 1)     #axis는 축의 방향이다. 1=열
df_seoul

```

```python
# '전입지별' 을 '전입지' 로 변경하기. 

df_seoul=df_seoul.rename(columns= {'전입지별': '전입지'})
df_seoul.set_index('전입지', inplace= True)
```

```python

sr_one= df_seoul.loc['경기도']
sr_one
```

```python
sr_one.head()
```

```python
sr_one.plot()
```

```python
plt.plot(sr_one.index, sr_one.values)
```

```python
plt.figure(figsize= (20,15))
plt.xticks(rotation= 60)
plt.yticks(rotation= 45)

plt.plot(sr_one)
plt.title('서울-> 경기 인구 이동')

plt.xlabel('기간')
plt.ylabel('인구 이동 수 ')

plt.legend(loc='left upper', labels=['서울-> 경기 '])

plt. saveifg("../Data/서울경기이동.png")

plt.show()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
#Style 서식 지정

plt.style.use('fast')
# 그림 사이즈 지정(가로 : 14인치, 세로 5인치 )
plt.figure(figsize=(14,5))


# X 축 눈금 라벨 회전하기 

plt.xticks(rotation=60)


# Y 축 눈금 라벨 회전하기

plt.yticks(rotation=60)


# marker 표시



# 차트 제목 추가

plt.plot(BS, marker="o", markersize=10)

plt.title('부산->서울 인구 이동')


# 축 이름 추가 

plt.xlabel('기간')
plt.ylabel('인구 이동 수 ')


plt.legend(labels= ['부산->서울'], loc= 'best')


# 그래프를 저장

plt.savefig("../Data/seoul.png")


plt.show()    # 변경사항 저장하고 그래프 출력  -> 메모리에서는 지워진다. 
```

```python
plt.style.use('ggplot')

plt.figure(figsize= (20,7))

plt.xlabel('기간')
plt.ylabel('인구 이동 수 ')

plt.legend(labels= ['서울-> 경기'], loc= 'upper right', fontsize= 10)

plt.xticks(rotation= 60, size =15)
plt.yticks(rotation='horizontal', size =15)

plt.plot(sr_one, marker='o', markersize= 10)
plt.title('서울-> 경기 이동 수 ')


plt.annotate(' ',
            xy= (20,630000),    #화살표의 머리 부분 (끝 점)
             xytext= (2,290000),   # 화살표의 꼬리 부분(시작 점)
             xycoords='data' ,   #좌표 체계
             arrowprops=dict(arrowstyle="->", color ='olive', lw=5)   #화살표 서식
            )
plt.annotate('인구이동 증가(1970~1995)',   #텍스트 입력
             xy= (10,400000),    #텍스트의 위치 기준점
             rotation = 22, #텍스트의 회전 각도 
            va ='baseline', #텍스트의 상하 정렬
             ha= 'center', # 텍스트의 좌우 정렬 
             fontsize =12   # 텍스트의 크기 
                         )


plt.annotate(' ',
            xy=(47,450000),
            xytext =(30,580000),
            xycoords='data',
            arrowprops=dict(arrowstyle="->", color='blue', lw=5))

plt.annotate('인구 이동 감소(1995~2017)',   
             xy= (40,500000),    
             rotation = -10,  
            va ='baseline', 
             ha= 'center',  
             fontsize =12  )        

    
    
plt.ylim(5000,800000)

plt.show()
```
